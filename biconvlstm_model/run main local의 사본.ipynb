```python
import glob
import os
import cv2
import numpy as np
```

```python
def read_video(filename):
    frames = []
    cap = cv2.VideoCapture(filename)
    while(cap.isOpened()):
        ret, frame = cap.read()  
        if not ret:
            break
        frames.append(frame)
        #print(np.array(frame).shape)
        break
    cap.release()
    print(filename, '\n',np.array(frames).shape)
    if np.array(frames).shape[0] == 0:
        return -1
    #print(video.shape)
    return 1
```

```python
root_dir = "/content/gdrive/Shareddrives/2021청년인재_고려대과정_10조/Test Data/AIHub_Dataset/fight"
data_files = glob.glob(root_dir + '/**/*.mp4', recursive=True)
```

```python
cnt = 0
for idx,i in enumerate(data_files):
    print('{}% done, {} out of {}'.format(round((idx+1)/len(data_files),2),idx+1,len(data_files)))
    videodata = read_video(i)
    if videodata == -1:
        #break
        try:
            os.remove(i)
        except:
            pass
        cnt+=1

print('finished and removed {} files'.format(cnt))
```
